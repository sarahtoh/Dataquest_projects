## Dealing with the columns with dates 

There are 5 columns that should represent date values. Some of these columns were created by the crawler, some came from the website itself. We can differentiate by referring to the data dictionary:

- `date_crawled`: added by the crawler
- `last_seen`: added by the crawler
- `ad_created`: from the website
- `registration_month`: from the website
- `registration_year`: from the website

Right now, the date_crawled, last_seen, and ad_created columns are all identified as string values by pandas, with full timestamp values. We will use the first 10 characters (the day) to calculate the distribution of values in the date_crawled, ad_created, and last_seen columns (all string columns) as percentages.

```python
# Calculating the distribution of values in date_crawled column
autos_filtered_price['date_crawled'].str[:10].value_counts(normalize=True, dropna=False).sort_index(ascending=False)
```

The data was scraped from the web between 5 Mar 2016 and 7 Apr 2016, with roughly equal amounts of entries scraped daily from the start date to 4 Apr 2016.

```python
autos_filtered_price['ad_created'].str[:10].value_counts(normalize=True, dropna=False).sort_index(ascending=False)
```

The ads were created between 11 Nov 2015 and 7 Apr 2016, with a larger % of ads created daily in the later part of this period.

```python
autos_filtered_price['last_seen'].str[:10].value_counts(normalize=True, dropna=False).sort_index(ascending=False)
```

The ads were last seen between 5 Mar 2016 to 7 Apr 2016, which is the same range of dates as all the scraping. However, the majority (~48%) were last seen on the last 3 days of scraping (5-7 Apr 2016), which suggests that nearly 50% of the total number of cars had not been sold by that date.

```python
autos_filtered_price['registration_year'].describe()
```

The mean year of registration is 2004, which makes is relatively recent for used cars. However, the minimum and maximum values for the registration year are not realistic (1000 and 9996), and should be removed as cars had not been invented in the former year and the latter has not passed yet.

Because a car can't be first registered after the listing was seen, any vehicle with a registration year above 2016 is definitely inaccurate, so that will be our upper bound. We will use 1920 as a generous lower bound, as it is after WWI and around the start of the period of tremendous growth in automobile manufacturing in Europe [link](https://www.britannica.com/technology/automotive-industry/Growth-in-Europe).

```python
# Filtering out impossible registration years and impossible power for the car
autos_filtered_year = autos_filtered_price[autos_filtered_price["registration_year"].between(1920,2016)]
autos_filtered_year = autos_filtered_year[autos_filtered_year['power_p_s'] < 1000]
autos_filtered_year['registration_year'].value_counts(normalize=True).sort_index(ascending=False).head(10)
```

After filtering out cars registered before 1920 and after 2016, we can see that there were a very small number of cars registered sporadically before the 1980s, and most cars were registered from the 1990s onwards.

## Aggregating data by brand for comparison

```python
# Getting the number of occurrences of each brand in the dataset as percentages
brand_counts = autos_filtered_year['brand'].value_counts(normalize=True).sort_values(ascending=False)
brand_counts
```

We can see that there are over 40 brands in the dataset and that most of the cars are concentrated among a small number of brands. Let's focus on brands that represent at least 5% of the cars in the dataset, which make up about 65% of the cars, to see how the average price compares across these brands.

```python
# Filtering the series to get only the most popular brands
top_brand = brand_counts[brand_counts >= 0.05]
top_brands = list(top_brand.index.values)
top_brands
```

```python
# Converting to dataframe to filter
# brand_df = brand_counts.to_frame(name='listings')
# top_brand = brand_df[brand_df['listings'] >= 0.05]
# top_brand = top_brand.reset_index()
# top_brand
```

```python
brands = {}

for brand in top_brands:
    prices = autos_filtered_year[autos_filtered_year['brand'] == brand]
    mean_price = prices["price"].mean()
    brands[brand] = mean_price
    
print(brands)
    
```

Of the 6 most popular brands for used cars listed on German eBay, 5 are German and one is America (Ford). Audi, Mercendes Benz and BMW are the most expensive, averaging 8000-9000 per car, while Opel and Ford are the cheapest at between 3000-4000 per car. Volkswagen is in the middle at about 5500 per car.

## Getting the average mileage per car for the top 6 brands

Let's use aggregation to understand the average mileage for those cars and if there's any visible link with mean price. While our natural instinct may be to display both aggregated series objects and visually compare them, this has a few limitations:

- It's difficult to compare more than two aggregate series objects if we want to extend to more columns
- We can't compare more than a few rows from each series object
- We can only sort by the index (brand name) of both series objects so we can easily make visual comparisons

Instead, we will create a dictionary for mean mileage too and combine all the values into a single dataframe for comparison.

```python
brands_mileage = {}

for brand in top_brands:
    mileage = autos_filtered_year[autos_filtered_year['brand'] == brand]
    mean_mileage = mileage["odometer_km"].mean()
    brands_mileage[brand] = mean_mileage
    
print(brands_mileage)
```

```python
mean_price_df = pd.DataFrame.from_dict(brands, orient='index', columns=['mean_price'])
# mean_price_df
```

```python
mean_mileage_series = pd.Series(brands_mileage)
mean_mileage_list = list(mean_mileage_series.values)
mean_price_df['mean_mileage'] = mean_mileage_list
# mean_mileage_series
mean_price_df
```

There is no visible link between mileage and price. BMW, Mercedes Benz and Audi cars have the highest average mileage (>129,000km), but are also teh most expensive. Furthermore, while Audi cars have similar average mileage to Volkswagen and Opel cars, the average price of Audis is significanly higher. Ford cars have the lowest average mileage by a bit (~120,000km) but are only the 2nd cheapest on average.
