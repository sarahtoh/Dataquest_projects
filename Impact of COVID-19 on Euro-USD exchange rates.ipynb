# COVID-19 did not have a lasting impact on Euro-USD exchange rates

For this project, we'll use a dataset describing Euro daily exchange rates between 1999 and 2021. The euro (symbolized with €) is the official currency in most of the countries of the European Union. The dataset was put together by Daria Chemkaeva and is available on [Kaggle](https://www.kaggle.com/lsind18/euro-exchange-daily-rates-19992020). The data source is the European Central Bank and gets regular updates. The data that we'll be visualising was downloaded on January 2021.

To start with, we'll learn some basic facts about the dataset.

```python
import pandas as pd
```

```python
exchange_rates = pd.read_csv('euro-daily-hist_1999_2020.csv')
```

```python
exchange_rates.head(5)
```

```python
exchange_rates.tail(5)
```

```python
exchange_rates.info()
```

We can see that the dataset contains the exchange rates of 40 currencies with the Euro, and that the currencies that were eventually replaced by the Euro have many null values. The data runs from 4 Jan 1999 to 8 Jan 2021.

## Focusing on Euro-USD exchange rates

```python
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar',
                               'Period\\Unit:': 'Time'},
                      inplace=True)
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
exchange_rates.sort_values('Time', inplace=True)
exchange_rates.reset_index(drop=True, inplace=True)
exchange_rates
```

As we want to zoom in specifically to Euro-USD exchange rates, we'll retain only the USD and date columns and do some cleaning to ensure that we only retain numerical values in the USD column.

```python
euro_to_dollar = exchange_rates[['Time','US_dollar']]
euro_to_dollar['US_dollar'].value_counts()
```

```python
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != '-']
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)
euro_to_dollar
```

### Visualising the exchange rates

```python
import matplotlib.pyplot as plt
%matplotlib inline
# Enables Jupyter to display graphs

plt.plot(euro_to_dollar['Time'],
         euro_to_dollar['US_dollar'])
plt.show()
```

The line is not smooth due to daily variations in the exchange rate with clear upward or downward trends only evident in the longer run (months or years).

If we want to focus only on the long-term trends, we can use the rolling mean to smoothen out the fluctuations.

```python
euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar'].rolling(30).mean()
```

### Looking at how the euro-USD rate changed during the COVID-19 pandemic

The COVID-19 pandemic impacted the global economy and had different impacts in different parts of the world, so it'll be interesting to see how it affected the Euro-USD exchange rate. 

To do this, we'll compare 2016-2019 data with 2020 data.

```python
import datetime as dt
import matplotlib.style as style
style.use('fivethirtyeight')
```

```python
euro_to_dollar['year'] = euro_to_dollar['Time'].dt.year
euro_to_dollar = euro_to_dollar[euro_to_dollar['year'] > 2015]
pre_covid = euro_to_dollar[euro_to_dollar['year'] < 2020]
covid = euro_to_dollar[euro_to_dollar['year'] > 2019]
euro_to_dollar
```

```python
average_rate = euro_to_dollar['US_dollar'].mean()
average_rate
covid_min = covid['US_dollar'].min()
covid_min
```

```python
lowest = covid[covid['US_dollar'] == covid_min]
lowest
```

```python
fig, ax = plt.subplots(figsize=(9, 5))
ax.plot(euro_to_dollar['Time'], euro_to_dollar['rolling_mean'],color='mediumblue', linewidth = 3,alpha = 0.3)
ax.plot(covid['Time'], covid['rolling_mean'], color='#af0b1e', linewidth = 3)

ax.grid(b=False)
ax.tick_params(bottom=0, left=0) 
# ax.set_yticklabels([])
# ax.set_xticklabels([])

for location in ['left','right','bottom','top']: 
    ax.spines[location].set_visible(False)

# ax.axvline(dt.datetime(2020, 1, 1), c='dimgrey', alpha=0.3, linewidth=1, ymin=0, ymax=1.5)
ax.axhline(average_rate, c='dimgrey', alpha=0.3, linewidth=1, xmin=0, xmax=0.95)
ax.axvline(dt.datetime(2020, 7, 5), c='dimgrey', alpha=0.3, linewidth=105, ymin=0, ymax=1.5)

ax.text(0.01, 0.48, 'Average rate: '+str(round(average_rate,2)), transform=ax.transAxes, color='grey')
ax.text(-0.1, 1.1, 'Euro-USD rate rebounded quickly from during COVID-19 pandemic', transform=ax.transAxes, size=16, weight='bold')
ax.text(-0.1, 1.03, 'Euro-USD exchanges rates from 2016-2020', transform=ax.transAxes) 
ax.text(-0.1, -0.15, '©DATAQUEST' + ' '*80 + 'European Central Bank', transform=ax.transAxes, color = '#f0f0f0', backgroundcolor = '#4d4d4d')

plt.show()
```
